```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
#Create a Line Chart Using the MATLAB Method
# Set the x-axis to a list of strings for each month.
x_axis =["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec" ]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.2, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09 ]
```

```python
# Create a Line Chart Using the Object-Oriented Interface
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot and addd a label for the legend
plt.plot(x_axis, y_axis, label='Boston')
```

```python
#Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
```

```python
# Set the y limit between 0 and 45
plt.ylim(0, 45)
```

```python
# Create a title.
plt.title("PyBer Fare by Month")
```

```python
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend.
plt.legend()
plt.figure(figsize=(20, 10))
```

```python
# Create a Line Chart Using the Object-Oriented Interface
# Create a the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis, marker="*", color="blue", linewidth=2, label="boston")
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0,45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid
ax.grid()
# Add the legend.
ax.legend()
fig, ax = plt.subplots(figsize=(20, 10))
```

```python
# Create a Line Chart Using the Object-Oriented Interface
# Create a the plot with ax.plt()
fig, ax = plt.subplots(figsize=(7,3.5))
ax.plot(x_axis,y_axis, marker="d", color="green", linewidth=2, label="boston")
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0,45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, color="red", label="Chicago",s=50)
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.legend()
```

```python
#Create a Bubble Chart Using the MATLAB Method
#We will make the size of each marker equal to the data floating-point decimal values of the y-axis by using s=y_axis.
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#We can adjust the size by multiplying the data in the y-axis by any value.  x 3
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Refactor the code we used to create a scatter plot above, but instead of using a for loop to create the list y_axis_larger for the size parameter of the marker - Use Phthon;s list comprehension techinque
# You can use list comprehension to replace many for and while loops. 
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, linewidth=2, color="skyblue", s=[i*3 for i in y_axis], label="Boston", alpha=0.50, edgecolor="black")
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.legend()
# Make the color 50% transparent.
# alpha=0.5
# Add a black edge color to the circles.
# edgecolor = "black"
# Make the linewidth of the circles 2 points.
# linewidth=2
# add a limit for the x axis
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# To pop out the seventh month, July, we use a decimal-point value of 0.2.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%1.1f%%")
```

```python
# The percentages on the pie chart seem a bit crowded. Let's increase the size of the pie chart by using plt.subplots(figsize=(8, 8)).
# To pop out the seventh month, July, we use a decimal-point value of 0.2.
# # Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
plt.show()
```
