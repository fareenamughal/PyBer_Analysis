```python
# Add Error Bars to a Line Chart
# plt.errorbar(x, y, yerr=<value>)
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Next, add the x_axis, y_axis, and yerr=stdev to the errorbar function:
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# Add a "cap" to the error bars that's equal to a scalar value. use capsize=
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Get the mean of the values in the y-axis.
mean = statistics.mean(y_axis)
mean
```

```python
# Add a dotted line one std dev above and below the mean 
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.axhline(y=stdev, color = "r", linestyle="dashed" )
plt.axhline(y=stdev+mean, color = "y", linestyle="dotted")
```

```python
# add error bars and a capsize using the object-oriented approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Add Error Bars to a Bar Chart using the MATLAB approach
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add Error Bars to a Bar Chart using the object-oriented approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add Error Bars to a Horizontal Bar Chart using the MATLAB approach & adjust major x-axis tiks
plt.barh(x_axis, y_axis, xerr=stdev, capsize=3)
```

```python
# To adjust the major x-axis ticks on a horizontal bar chart, we use the xticks() function.

# Add the following code to a new cell:

import numpy as np
plt.barh(x_axis, y_axis, xerr=stdev, capsize=3)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
import numpy as np
np.arange(0, 5, step=5.0)
plt.barh(x_axis, y_axis, xerr=stdev, capsize=3)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Add Error Bars to a Horizontal Bar Chart using the bject-oriented approach & adjust major x-axis tiks
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
ax.invert_yaxis()
```

```python
# Add minor ticks to the x-axis on a horizontal bar chart, we use the set_minor_locator function with the xaxis class in the object-oriented method
# Inside the set_minor_locator function, we add the MultipleLocator method and pass how often we want our minor ticks.

# To use these functions and methods, we have to import the MultipleLocator method from the matplotlib.ticker function.
# x.axis.set_minor_locator()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
